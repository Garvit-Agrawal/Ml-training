```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

##### Categories
1. Catagorical Plots
    a.countplot
    b.barplot
    c.boxplot
    d.violinplot
    e.swarmplot
2. Distribution Plots
    a.distplot
    b.displot
    c.jointplot
    d.pairplot
    e.kdeplot
    f.rugplot
    g.scatterplot
3. Matrix Plots
    a.heatmap
    b.clustermap
    c.pairplot(same as above one)
4. Regression Plots
    a.lmplot
    b.regplot
5. Grid Plots
    a.PairGrid
    b.JointGrid
    c.FacetGrid

```python
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.set_style("whitegrid")
```

```python
sns.distplot(tips["total_bill"])
```

```python
sns.displot(data=tips,x="total_bill")
```

```python
sns.displot(data=tips,x="total_bill",y="tip")
```

```python
sns.jointplot(data=tips,x="total_bill",y="tip");plt.show()
sns.jointplot(data=tips,x="total_bill",y="tip",hue="smoker");plt.show()
sns.jointplot(data=tips,x="total_bill",y="tip",kind="hex");plt.show()
```

```python
sns.scatterplot(data=tips,x="total_bill",y="tip")
```

```python
fig=plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
sns.kdeplot(data=tips,x="total_bill")
sns.rugplot(data=tips,x="total_bill")
plt.subplot(1,2,2)
sns.kdeplot(data=tips,x="total_bill",y="tip")
sns.rugplot(data=tips,x="total_bill",y="tip")
```

```python
iris=sns.load_dataset("iris")
```

```python
sns.pairplot(data=iris)
```

```python
sns.pairplot(data=iris,hue="species")
```

```python
sns.regplot(data=tips,x="total_bill",y="tip");plt.show()
sns.lmplot(data=tips,x="total_bill",y="tip");plt.show()
```

```python
pg=sns.PairGrid(data=iris)
pg.map_diag(sns.kdeplot)
pg.map_lower(sns.scatterplot)
pg.map_upper(sns.kdeplot)
```

```python
jeg=sns.JointGrid(data=tips,x="total_bill",y="tip")
```

```python
jg=sns.JointGrid(data=tips,x="total_bill",y="tip")
jg.plot(sns.regplot,sns.distplot)
```

```python
fg=sns.FacetGrid(data=tips,row="smoker",col="day",hue="sex")
fg.map(sns.scatterplot,"total_bill","tip").add_legend()
```
